**DATA CLEANING USING PANDAS**

Raw datasets often contain inconsistencies that hinder analysis. The project resolves these using Pandas:
Handling Missing Values: Replace missing numerical values with defaults like 0 or drop rows with critical missing fields.
Renaming Columns: Standardize column names for clarity and compatibility with SQL databases (e.g., converting Order ID to order_id).

```python
import pandas as pd
df = pd.read_csv("/kaggle/input/retail-orders/orders.csv") #reads file from pc

#Corrects the headings
def clean_column_names(df): #to remove trailing spaces
    df.rename(columns=lambda x: str(x).replace(" ", "_").lower(), inplace=True)

#To change column name
old_names = list(df.keys())
updated_names = [i.replace(" ", "_") for i in old_names]
updated_names = [j.lower() for j in updated_names]
df.columns = updated_names

#Adding new columns

# Step 1: Calculate Discount
df['discount'] = df['list_price'] * (df['discount_percent'] / 100)

# Step 2: Calculate Sale Price
df['sale_price'] = df['list_price'] - df['discount']

# Step 3: Calculate Profit
df['profit'] = df['sale_price'] - df['cost_price']


output_file_path = 'updated_file.csv'  #saving it in new file
df.to_csv(output_file_path, index=False)

#Downloading it into PC
from google.colab import files #Downloading it into PC
files.download('updated_file.csv')
```

```python
%%writefile app.py
import streamlit as st
import pymysql
import pandas as pd

connection = pymysql.connect(
  host = "gateway01.ap-southeast-1.prod.aws.tidbcloud.com",
  port = 4000,
  user = "3HPcjPsRYP9Wwei.root",
  password = "jQW1qU4bMWZVI4gh",
  database = "A3",
  ssl_verify_cert = True,
  ssl_verify_identity = True
)
mycursor=connection.cursor()
mycursor.execute("USE kaggle_project")


# Create a cursor
mycursor = connection.cursor()

# Function to fetch data and return as DataFrame
def fetch_data(query):
    mycursor.execute(query)
    out = mycursor.fetchall()
    column_names = [desc[0] for desc in mycursor.description]
    df = pd.DataFrame(out, columns=column_names)
    df.insert(0, "Serial No", range(1, len(df) + 1))
    return df

# Dictionary containing all queries with insights
queries = {
    "1. Top-Selling Products": {
        "query": """
            SELECT sub_category, category, SUM(quantity * sale_price) AS total_revenue
            FROM kaggle_project.kaggle
            GROUP BY sub_category, category
            ORDER BY total_revenue DESC
            LIMIT 10
        """,
        "insights": [
            "These products generate the highest revenue, indicating strong market demand.",
            "Pair slow-moving products with high-revenue items at a discount to maximize sales."
        ]
    },
    "2. Monthly Sales Analysis": {
        "query": """
            SELECT EXTRACT(YEAR FROM order_date) AS year, EXTRACT(MONTH FROM order_date) AS month,
            SUM(quantity * sale_price) AS total_revenue
            FROM kaggle_project.kaggle
            GROUP BY year, month
            ORDER BY year, month
        """,
        "insights": [
            "Focus on months with declining sales to implement targeted promotions or discounts.",
            "Introduce mid-year promotions, back-to-school campaigns, or limited-time summer discounts."
        ]
    },
    "3. Highest Sales in Year and Month": {
        "query": """
            WITH MonthlyRevenue AS (
                SELECT YEAR(order_date) AS year, MONTH(order_date) AS month,
                SUM(quantity * sale_price) AS total_revenue
                FROM kaggle_project.kaggle
                GROUP BY year, month
            ),
            MoMRevenue AS (
                SELECT year, month, total_revenue,
                LAG(total_revenue, 1, 0) OVER (ORDER BY year, month) AS previous_month_revenue
                FROM MonthlyRevenue
            )
            SELECT year, month, total_revenue, previous_month_revenue,
            (total_revenue - previous_month_revenue) AS revenue_change
            FROM MoMRevenue
            ORDER BY year, month
        """,
        "insights": [
            "Identify seasonal trends and sales growth or decline.",
            "Focus on months with declining sales for targeted strategies."
        ]
    },
    "4. Lowest Monthly and Yearly Sales": {
        "query": """
            SELECT EXTRACT(YEAR FROM order_date) AS year, EXTRACT(MONTH FROM order_date) AS month,
            SUM(quantity * sale_price) AS total_revenue
            FROM kaggle_project.kaggle
            GROUP BY year, month
            ORDER BY total_revenue ASC
            LIMIT 1
        """,
        "insights": [
            "Festive seasons and cheaper alternatives may lead to lower sales.",
            "Compare low-sales months with high-sales months to improve overall performance."
        ]
    },
    "5. Revenue per Category": {
        "query": """
            SELECT category, sub_category, SUM(quantity * sale_price) AS total_revenue,
            CASE
                WHEN SUM(quantity * sale_price) > 50000 THEN 'High Revenue'
                WHEN SUM(quantity * sale_price) BETWEEN 20000 AND 50000 THEN 'Medium Revenue'
                ELSE 'Low Revenue'
            END AS revenue_category
            FROM kaggle_project.kaggle
            GROUP BY category, sub_category
            ORDER BY sub_category, total_revenue DESC
        """,
        "insights": [
            "Identify which product categories generate the highest and lowest revenue.",
            "Use this data to allocate resources effectively and prioritize marketing efforts."
        ]
    },
    "6. High-Profit Products": {
        "query": """
            SELECT
                sub_category,
                SUM((sale_price - cost_price) * quantity) AS total_profit,
                SUM(quantity * sale_price) AS total_revenue,
                (SUM((sale_price - cost_price) * quantity) / SUM(quantity * sale_price)) * 100 AS profit_margin
            FROM kaggle_project.kaggle
            WHERE quantity > 0 AND sale_price > 0
            GROUP BY sub_category
            HAVING profit_margin > 0.1
        """,
        "insights": [
            "Identify products with the highest profit margins to focus on maximizing profitability.",
            "Analyze the factors contributing to high profit margins (e.g., pricing, cost efficiency, or high demand).",
            "Use this data to replicate successful strategies in other sub-categories or products."
        ]
    },
    "7. Lowest Profit Products": {
        "query": """
            SELECT
                sub_category,
                SUM((sale_price - cost_price) * quantity) AS total_profit,
                SUM(quantity * sale_price) AS total_revenue,
                (SUM((sale_price - cost_price) * quantity) / SUM(quantity * sale_price)) * 100 AS profit_margin
            FROM kaggle_project.kaggle
            WHERE quantity > 0 AND sale_price > 0
            GROUP BY sub_category
            HAVING profit_margin > 0
            LIMIT 10
        """,
        "insights": [
            "Identify products with the lowest profit margin and improve the quality to maximize profitability.",
            "Reduce production costs and adjust the number of units to align with sales expectations."
        ]
    },
    "8. Regional Sales Analysis": {
        "query": """
            SELECT
                region,
                SUM(quantity * sale_price) AS total_revenue,
                SUM(quantity) AS total_units_sold
            FROM kaggle_project.kaggle
            GROUP BY region
            ORDER BY total_revenue DESC
        """,
        "insights": [
            "Identify regions with the highest sales to focus on expanding operations or increasing marketing efforts.",
            "Analyze underperforming regions to determine potential causes (e.g., lack of awareness, competition, or logistics issues).",
            "Tailor regional strategies based on sales performance to maximize overall revenue."
        ]
    },
    "9. Products with Discounts Greater than 20%": {
        "query": """
            SELECT
                category,
                sub_category AS discounted_products,
                SUM(quantity) AS total_units_sold,
                SUM(sale_price * quantity) AS total_revenue,
                SUM(discount) AS total_discount_given,
                ROUND((SUM(discount) / NULLIF(SUM(sale_price * quantity), 0)) * 100, 2) AS avg_discount_percent
            FROM kaggle_project.kaggle
            WHERE discount_percent > 0.2
            GROUP BY category, sub_category
            ORDER BY total_discount_given DESC
        """,
        "insights": [
            "Identify products with high discounts and analyze their impact on revenue and profitability.",
            "Determine if high discounts are driving sufficient sales volume or eroding profit margins."
        ]
    },
    "10. Impact of Discount on Sales": {
        "query": """
            SELECT
                sub_category,
                category,
                SUM(sale_price * quantity) AS total_revenue
            FROM kaggle_project.kaggle
            GROUP BY sub_category, category
            ORDER BY total_revenue DESC
            LIMIT 10
        """,
        "insights": [
            "Discounts can boost sales by attracting price-sensitive customers, increasing short-term revenue.",
            "Frequent discounts may reduce profit margins and lower perceived product value over time."
        ]
    },
"11. Top 10 Highest Revenue Generating Products": {
        "query": """
            SELECT category, sub_category, SUM(sale_price * quantity) AS total_revenue
            FROM kaggle_project.kaggle
            GROUP BY category, sub_category
            ORDER BY total_revenue DESC
            LIMIT 10
        """,
        "insights": [
            "1. Identifies the most profitable products, enabling better inventory and marketing focus.",
            "2. Focus on these products to maximize revenue and ensure consistent stock availability."
        ]
    },
    "12. Top 5 Cities with the Highest Profit Margin": {
        "query": """
            SELECT city, SUM(profit) AS total_profit, SUM(sale_price * quantity) AS total_revenue,
                   ROUND((SUM(profit) / SUM(sale_price * quantity)) * 100, 2) AS profit_margin
            FROM kaggle_project.kaggle
            GROUP BY city
            ORDER BY profit_margin DESC
            LIMIT 5
        """,
        "insights": [
            "1. Identifies cities with the most efficient operations or highest customer willingness to pay.",
            "2. Focus on these cities for expansion or targeted marketing campaigns to replicate success."
        ]
    },
    "13. Total Discount Given for Each Category": {
        "query": """
            SELECT category, sub_category AS discounted_products, SUM(quantity) AS total_units_sold,
                   SUM(sale_price * quantity) AS total_revenue, SUM(discount) AS total_discount_given,
                   ROUND((SUM(discount) / NULLIF(SUM(sale_price * quantity), 0)) * 100, 2) AS avg_discount_percent
            FROM kaggle_project.kaggle
            WHERE discount_percent > 0.2
            GROUP BY category, sub_category
            ORDER BY total_discount_given DESC
        """,
        "insights": [
            "1. Identifies categories with excessive discounts, which may impact profitability.",
            "2. Consider reducing unnecessary discounts in high-discount categories to improve profit margins."
        ]
    },
    "14. Average Sale Price per Product Category": {
        "query": """
            SELECT category, sub_category, ROUND(AVG(sale_price), 2) AS avg_sale_price
            FROM kaggle_project.kaggle
            GROUP BY category, sub_category
            ORDER BY avg_sale_price DESC
        """,
        "insights": [
            "1. Understanding the average sale price per category helps in pricing strategy and competitive positioning.",
            "2. Lower-priced categories may require upselling or bundling strategies to increase revenue."
        ]
    },
    "15. Region with the Highest Average Sale Price": {
        "query": """
            SELECT category, region, ROUND(AVG(sale_price), 2) AS avg_sale_price
            FROM kaggle_project.kaggle
            GROUP BY category, region
            ORDER BY avg_sale_price DESC
            LIMIT 5
        """,
        "insights": [
            "1. Identifies regions where customers are willing to pay premium prices.",
            "2. Target these regions for premium product launches and focused marketing efforts."
        ]
    },
    "16. Total Profit per Category": {
        "query": """
            SELECT category, SUM((sale_price - cost_price) * quantity) AS total_profit
            FROM kaggle_project.kaggle
            GROUP BY category
            ORDER BY total_profit DESC
        """,
        "insights": [
            "1. Reveals which categories contribute the most to overall profitability.",
            "2. Optimize costs or discontinue underperforming products in low-profit categories."
        ]
    },
    "17. Top 3 Segments with the Highest Quantity of Orders": {
        "query": """
            SELECT segment, SUM(quantity) AS total_quantity
            FROM kaggle_project.kaggle
            GROUP BY segment
            ORDER BY total_quantity DESC
            LIMIT 3
        """,
        "insights": [
            "1. Highlights the most active customer segments, guiding marketing efforts.",
            "2. Focus on these segments to enhance loyalty and engagement."
        ]
    },
    "18. Average Discount Percentage per Region": {
        "query": """
            SELECT region, ROUND(AVG(discount_percent), 2) AS avg_discount_percent
            FROM kaggle_project.kaggle
            GROUP BY region
            ORDER BY avg_discount_percent DESC
        """,
        "insights": [
            "1. Reveals regions where discounts are heavily used, affecting profitability.",
            "2. Adjust pricing strategies in high-discount regions to improve margins."
        ]
    },
    "19. Product Category with the Highest Total Profit": {
        "query": """
            SELECT category, SUM((sale_price - cost_price) * quantity) AS total_profit
            FROM kaggle_project.kaggle
            GROUP BY category
            ORDER BY total_profit DESC
            LIMIT 1
        """,
        "insights": [
            "1. Identifies the most profitable category, a key driver of business success.",
            "2. Maintain and expand this category to sustain long-term profitability."
        ]
    },
    "20. Total Revenue Generated per Year": {
        "query": """
            SELECT EXTRACT(YEAR FROM order_date) AS year, SUM(sale_price * quantity) AS total_revenue
            FROM kaggle_project.kaggle
            GROUP BY year
            ORDER BY year
        """,
        "insights": [
            "1. Tracks overall business performance and revenue trends over time.",
            "2. Investigate declining revenue years to identify causes and take corrective actions."
        ]
    }
}

# Streamlit UI
st.title("Retail Order Data Analysis ")
st.write("Select a query from the dropdown to view data and insights.")

# Dropdown to select query
selected_query = st.selectbox("Choose a Query", list(queries.keys())) #streamlit function

# Fetch and display data
if selected_query:
    query_info = queries[selected_query]
    data = fetch_data(query_info["query"])
    st.write(f"## {selected_query}")
    st.table(data)

    # Display Insights
    st.write("### Insights:")
    for insight in query_info["insights"]:
        st.write(f"- {insight}")

# Close the cursor and connection
mycursor.close()
connection.close()
```
